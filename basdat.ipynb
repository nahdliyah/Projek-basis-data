```python
import pandas as pd
import random
import string

# Fungsi untuk membuat kode unik
def generate_kode_unik(huruf_stasiun):
    angka_acak = ''.join(random.choices(string.digits, k=4))
    kode_baru = huruf_stasiun + angka_acak
    return kode_baru

# # Baca file CSV
# file_path = 'path/to/your/file.csv'
# data = pd.read_csv(file_path)

# # Hapus baris dengan nilai kosong
# data_tanpa_kosong = data.dropna()

# Tambahkan kolom 'kode_unik' berdasarkan 'kode stasiun'
data_tanpa_kosong['kode_unik'] = data_tanpa_kosong['fcode'].apply(lambda x: generate_kode_unik(x[:]))

# Tampilkan hasil
print(data_tanpa_kosong[['fcode', 'kode_unik']])
```

```python
import pandas as pd
import random

# Fungsi untuk menghasilkan jam acak dengan selisih antara waktu berangkat dan datang
def generate_jam_acak_berangkat():
    jam = random.randint(0, 23)
    menit = random.randint(0, 59)
    return f"{jam:02d}:{menit:02d}"

def generate_jam_acak_datang(berangkat):
    jam_berangkat = int(berangkat.split(':')[0])
    menit_berangkat = int(berangkat.split(':')[1])

    selisih_jam = random.randint(0, 11)  # Selisih maksimal 11 jam
    selisih_menit = random.randint(2, 59)  # Selisih minimal 2 menit dan maksimal 59 menit

    jam_datang = (jam_berangkat + selisih_jam) % 24
    menit_datang = (menit_berangkat + selisih_menit) % 60

    return f"{jam_datang:02d}:{menit_datang:02d}"

# Tambahkan kolom 'waktu_berangkat' dan 'waktu_datang'
data_tanpa_kosong['waktu_berangkat'] = [generate_jam_acak_berangkat() for _ in range(len(data_tanpa_kosong))]
data_tanpa_kosong['waktu_datang'] = data_tanpa_kosong['waktu_berangkat'].apply(generate_jam_acak_datang)

# Tampilkan hasil
print(data_tanpa_kosong)
```

```python
from datetime import datetime, timedelta
import pandas as pd
import random

# Fungsi untuk menghasilkan tanggal acak dalam rentang tertentu
def generate_tanggal_acak(start_date, end_date):
    start_datetime = datetime.strptime(start_date, '%m/%d/%Y')
    end_datetime = datetime.strptime(end_date, '%m/%d/%Y')
    
    days_difference = (end_datetime - start_datetime).days
    random_days = random.randint(0, days_difference)
    
    tanggal_acak = start_datetime + timedelta(days=random_days)
    return tanggal_acak.strftime('%m/%d/%Y')

# Tambahkan kolom 'tanggal'
start_date = '12/10/2023'
end_date = '01/05/2024'
data_tanpa_kosong['tanggal'] = [generate_tanggal_acak(start_date, end_date) for _ in range(len(data_tanpa_kosong))]

# Tampilkan hasil
print(data_tanpa_kosong)
```

```python
import pandas as pd
import random

# Daftar Nama Kereta Api
nama_kereta_api = [
    'Argo Bromo Anggrek', 'Argo Dwipangga', 'New Argo Jati', 'Argo Lawu', 
    'Argo Muria', 'Argo Parahyangan', 'Argo Sindoro', 'Argo Wilis',  
    'Bangunkarta', 'Bima', 'Gajayana', 'Sembrani', 'Taksaka', 'Turangga',
    'Argo Mawar', 'Gajayana', 'Muntilan', 'Jayakarta', 'Orchid Jaya'
]

# Fungsi untuk menghasilkan kelas kereta acak
def generate_kelas_acak():
    kelas_kereta = ['Ekonomi', 'Bisnis', 'Eksekutif', 'Premium', 'Pariwisata']
    return random.choice(kelas_kereta)

# Fungsi untuk memetakan kelas ke nama kereta, kapasitas, dan harga
def map_kelas_to_info(kelas):
    if kelas == 'Ekonomi':
        return random.choice(nama_kereta_api), 200, 100000
    elif kelas == 'Bisnis':
        return random.choice(nama_kereta_api), 150, 200000
    elif kelas == 'Eksekutif':
        return random.choice(nama_kereta_api), 100, 300000
    elif kelas == 'Premium':
        return random.choice(nama_kereta_api), 50, 400000
    elif kelas == 'Pariwisata':
        return random.choice(nama_kereta_api), 30, 500000


# Tambahkan kolom 'kelas'
data_tanpa_kosong['kelas'] = [generate_kelas_acak() for _ in range(len(data_tanpa_kosong))]

# Tambahkan kolom 'nama_kereta' dan 'kapasitas' berdasarkan kelas
data_tanpa_kosong[['nama_kereta', 'kapasitas',"harga"]] = pd.DataFrame(data_tanpa_kosong['kelas'].apply(map_kelas_to_info).tolist(), index=data_tanpa_kosong.index)
# Tampilkan hasil
print(data_tanpa_kosong)
```

```python
import pandas as pd
import random
import string

# Fungsi untuk membuat kode unik
def generate_kode_unik(huruf_stasiun, existing_codes, max_attempts=100):
    for _ in range(max_attempts):
        angka_acak = ''.join(random.choices(string.digits, k=4))
        kode_baru = huruf_stasiun + angka_acak
        if kode_baru not in existing_codes:
            existing_codes.add(kode_baru)  # Tambahkan ke set untuk memastikan keunikan
            return kode_baru
    raise ValueError(f"Tidak dapat menemukan kode unik yang unik setelah {max_attempts} percobaan.")


# Inisialisasi himpunan yang berisi kode unik yang sudah ada
existing_codes = set(data_tanpa_kosong['kode_unik'])

# Tambahkan kolom 'kode_unik' berdasarkan 'kode stasiun'
data_tanpa_kosong['kode_unik'] = data_tanpa_kosong.apply(lambda row: generate_kode_unik(row['fcode'], existing_codes), axis=1)

# Tampilkan hasil
print(data_tanpa_kosong[['fcode', 'kode_unik']])
```
